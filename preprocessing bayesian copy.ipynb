```python
def yaml(r,g,b):
    path2 = 'PolypsSet'
    x = str(r) + 'r' + str(g) + 'g' + str(b) + 'b'
    y = [1,2,3,4,5] # 5 fold
    for i in y:
        out_file = open(path2 + '/Imagenew3/direct_trainnew/' + x + '_polyp_yolov8_' +str(i)+'.yaml', 'w')
        out_file.write('train: PolypsSet/Imagenew3/Image_norm_'+ x + '/trainfold2/'+ str(i) + '/train' + '\n' + 'val: PolypsSet/Imagenew3/Image_norm_'+ x + '/' + 'val'  + '\n' + 'test: PolypsSet/Imagenew3/Image_norm_'+ x + '/trainfold2/'+ str(i) + '/val' + '\n' + 'nc: 2' + '\n' + "names: ['adenomatous','hyperplastic']")
        out_file.close()
```

```python
def black_box_function(r,g,b):
    r = round(r,1)
    g = round(g,1)
    b = round(b,1)
    prepro(r,g,b)
    yaml(r,g,b)
    sumx = 0
    y = [1,2,3,4,5]
    for fold in y:
        # print(sumx)
        i = str(r) + 'r' + str(g) + 'g' + str(b) + 'b_polyp_yolov8_'+ str(fold) + '.yaml'
        print('start traing '+i)
        j = 'PolypsSet/Imagenew3/direct_trainnew/' + i
        ###############
        train_n = os.listdir('runs/detect')
        m = []
        for detect in train_n:
            if len(detect) == 8:
                m.append(detect)
        o = m[-1]
        n = str(int(o[-3:])+1)
        #################
        cmd = "yolo task=detect mode=train model=yolov8s.pt data=%s epochs=50"%(j)
        print('___________________________________________________________________________________________________')
        !{cmd}
        print('---------- train ' + i + ' is done-------------' )
        ###################################################\n",
        k = 'runs/detect/train' + n + '/weights/best.pt'
        print(k)
        cmd = "yolo task=detect mode=val model=%s data=%s"%(k,j)
        output = !{cmd}
        all = output[-4]
        print(all)
        x = [i for i in all.split(" ") if i is not ''][-2]
        x = float(x)
        print(x)
        sumx = sumx + x
        print('############## detect/train'+ n +' ################')
        print('___________________________________________________________________________________________________')
    meanx=sumx/len(y)
    print(meanx)
    print('^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^')
    return meanx
```

```python
import shutil
from bayes_opt import BayesianOptimization
from bayes_opt.logger import JSONLogger
from bayes_opt.event import Events
```

```python
# Bounded region of parameter space
pbounds = {'r': (0, 2), 'g': (0, 2),'b': (0, 2)}

optimizer = BayesianOptimization(f=black_box_function,pbounds=pbounds,random_state=1)

logger = JSONLogger(path="./logs.json")
optimizer.subscribe(Events.OPTIMIZATION_STEP, logger)

optimizer.maximize(init_points=1000,n_iter=100)
```

```python
from bayes_opt.util import load_logs

new_optimizer = BayesianOptimization(
    f=black_box_function,
    pbounds = {'r': (0, 2), 'g': (0, 2),'b': (0, 2)},
    verbose=2,
    random_state=1,
)

# New optimizer is loaded with previously seen points
load_logs(new_optimizer, logs=["./logs.json"])
```
